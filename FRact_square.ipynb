```python
import turtle
from turtle import *
import time
from time import *
#speed
speed(0)
delay(0)
tracer(0,0)
#Initial Sequence 
list_0 = ["R","F","R","R","D2","F","D1","F","F","R","F","D2","F"]
# ["D1","R","R","D3","L","L","D1","R","D1","D4","L","L"]
#list_1 = ["L"] Useless bit of code
#Loop control
running = 2
#Distance between each turn
size = 2

itr=13;

#Angle of each turn
angle_1 = 85
#Angle for 2nd term
angle_2=180
#Last angle turn
last_angle = 180

#Canvas size 
screensize(2000,2000)
#color control
color_used = 1
colormode(255)
R_color = 1
G_color = 1
B_color = 1
RGB_color = [R_color, G_color, B_color]
#Hides turtle
hideturtle()
#Main Loop(Only Loop. lol)
for i in range(itr):
    time_0 = time()
    length = len(list_0)
    print("amount of iterations: ",itr)
    update()
    for inc in range(1,length):
        if color_used == 1:
            color((R_color, G_color, B_color))
#List decoder
        if list_0[inc] == "R":
            #print(list_0)
            right(90)
            forward(size)
        if list_0[inc] == "L":
            #print(list_0)
            left(90)
            forward(size)
        if list_0[inc] == "F":
            #print(list_0)
            #left(45)
            forward(4*size)
        if list_0[inc] == "D1":
            #print(list_0)
            left(45)
            forward(size)
        if list_0[inc] == "D2":
            #print(list_0)
            left(315)
            forward(size)
        if list_0[inc] == "D3":
            #print(list_0)
            left(225)
            forward(3*size)
        if list_0[inc] == "D4":
            #print(list_0)
            left(135)
            forward(size)
 
        R_color += 5
        if R_color >= 255:
            R_color = 1
            G_color += 5
            if G_color >= 255:
                G_color = 1
                B_color += 5
                     
    print("Current Color",R_color, G_color, B_color)
#Duplication and turning sequence             
    list_0.extend(list_0)
    right(last_angle)
    forward(size)
    list_0.append("R")
turtle.done()
#only use this if you want to see the operation that is in progress
    #print(list_0)
#     time_1 = time()
#     print(int(time_1 - time_0),"seconds")
```
